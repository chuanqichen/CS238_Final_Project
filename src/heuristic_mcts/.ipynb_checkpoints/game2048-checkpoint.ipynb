```julia
using DecisionMakingProblems
using MCTS
using Statistics
using Plots
using LaTeXStrings
using BenchmarkTools
using POMDPs
using POMDPModelTools
using MCTS
using POMDPSimulators
using POMDPPolicies
using BenchmarkTools
using Plots
using Statistics
```

```julia
using DecisionMakingProblems: TwentyFortyEight, isdone, get_max_rank, count_distinct_tiles, insert_tile_rand
using DecisionMakingProblems: initial_board, score_board, rollout_to_end, reward, MDP, print_board, draw_tile
using DecisionMakingProblems: init_2048_tables, move_down, move_left, move_right, move, count_empty, play_game
using DecisionMakingProblems: DIRECTIONS
```

```julia
using Game2048
```

```julia
include("./GM2048.jl")
include("./dm.jl")
```

```julia
mdp2048 = MDP(TwentyFortyEight(γ=0.99));
```

```julia
π_rand(s) = rand(DIRECTIONS)
play_game(π_rand)
```

```julia
rollouts_random = [ DM.rollout(mdp2048, initial_board(), π_rand, 10000, isdone) for i in 1:100 ]

print("π_rand: $(mean(rollouts_random)) ± $(std(rollouts_random))")
```

```julia
histogram(rollouts_random, label="Total Rewards for π_rand")
```

```julia
@benchmark DM.rollout(mdp2048, initial_board(), π_rand, 10000, isdone)
```

```julia
mcts_2048 = DM.MonteCarloTreeSearch(
    mdp2048, # 𝒫
    Dict{Tuple{UInt64, UInt8}, Float64}(), # Q
    Dict{Tuple{UInt64, UInt8}, Float64}(), # N
    10, # d
    100, # m
    5.0, # c
    s -> 0 # U
)

board2048 = initial_board();
```

```julia
rollouts_mcts = [ DM.rollout(mdp2048, initial_board(), mcts_2048, 10000, isdone) for i in 1:100 ]

print("mcts_2048: $(mean(rollouts_mcts)) ± $(std(rollouts_mcts))")
histogram(rollouts_mcts, label=L"\mathrm{MCTS} \quad U(s) = \mathbb{E}\,[R\,|\,\pi_{rand}]")
```

```julia
@benchmark DM.rollout(mdp2048, board2048, mcts_2048, 10000, isdone)
```

```julia
U(s) = 4 * count_empty(s) + 2* score_board(s) + score_board(s) >=2048 ? 50000 : 0

mcts_2048_U = DM.MonteCarloTreeSearch(
    mdp2048, # 𝒫
    Dict{Tuple{UInt64, UInt8}, Float64}(), # Q
    Dict{Tuple{UInt64, UInt8}, Float64}(), # N
    10, # d
    100, # m
    5.0, # c
    s -> U(s) # U(s) # U
)

rollouts_mcts_U = [ DM.rollout(mdp2048, initial_board(), mcts_2048_U, 1000, isdone) for i in 1:100 ]
print("mcts_2048_U: $(mean(rollouts_mcts_U)) ± $(std(rollouts_mcts_U))")
```

```julia
histogram(rollouts_mcts_U, label=L"\mathrm{MCTS} \quad U(s) Heuristic]")
```

```julia
play_game(s -> mcts_2048_U(s))
```
